# Comparison of measurements from different libraries

We now compare the implementation of the measurement `perimeter` in [napari-skimage-regionprops](https://github.com/haesleinhuepf/napari-skimage-regionprops) `regionprops_table` and [napari-simpleitk-image-processing](https://github.com/haesleinhuepf/napari-simpleitk-image-processing) `label_statistics`.

```python
import numpy as np
import pandas as pd
from skimage.io import imread
from pyclesperanto_prototype import imshow
import matplotlib.pyplot as plt
from napari_skimage_regionprops import regionprops_table
from napari_simpleitk_image_processing import watershed_otsu_labeling
from napari_simpleitk_image_processing import label_statistics
```

```python
blobs = imread('../../data/blobs.tif')
labels = watershed_otsu_labeling(blobs)
imshow(labels, labels=True)
```

```python
skimage_statistics = regionprops_table(blobs, labels, perimeter = True)
skimage_statistics
```

```python
simpleitk_statistics = label_statistics(blobs, labels, size = False, intensity = False, perimeter = True)
simpleitk_statistics
```

```python
simpleitk_perimeter = np.asarray(simpleitk_statistics['perimeter'])[1:]
```

```python
skimage_perimeter = np.asarray([s.feret_diameter_max for s in skimage_statistics])[1:]
```

## Scatter plot

```python
plt.plot(skimage_perimeter, simpleitk_perimeter, 'o')
```

## Bland-Altman plot

```python
mean = (skimage_perimeter + simpleitk_perimeter) / 2
diff = skimage_perimeter - simpleitk_perimeter

plt.plot(mean, diff, 'o')
```
